```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("Customer Churn.csv")
df.head()
```

```python
df.info()
```

#replacing blanks with 0 as tenure is 0 and no total charges are recoreded

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ","0")
df["TotalCharges"]=df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum().sum() # isnull() to find null values .sum() to find sum of null
```

```python
df["customerID"].duplicated().sum()
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "No"
df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

#convert 0 and 1 values of senior citizen to yes/no to make it easier to understand

```python
df.head(20)
```

```python
ax=sns.countplot(x='Churn',data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
plt.figure(figsize=(4,4))
gb=df.groupby('Churn').agg({'Churn':"count"})

plt.pie(gb['Churn'],labels=gb.index,autopct="%1.2f%%")
plt.title("Percentage of Churned Customers",fontsize=10)
plt.show()
```

#from the given pie chart we can conclude 26.45% of our customers have churned out
#now lets explore the reason behind it

```python
plt.figure(figsize=(4,4))
sns.countplot(x='gender',data=df,hue="Churn")
plt.title("Churn by gender")
plt.show()
```

```python
plt.figure(figsize=(3,3))
ax=sns.countplot(x='SeniorCitizen',data=df,hue="Churn")
ax.bar_label(ax.containers[0])
plt.title("Churn by Senior Citizen")
plt.show()
```

```python
grouped = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)
percentages = grouped.div(grouped.sum(axis=1), axis=0) * 100

# Plot
fig, ax = plt.subplots(figsize=(6,6))
bottom = None

for churn in percentages.columns:
    if bottom is None:
        bottom = percentages[churn]
        ax.bar(percentages.index, percentages[churn], label=churn)
    else:
        ax.bar(percentages.index, percentages[churn], bottom=bottom, label=churn)
        bottom += percentages[churn]

# Add percentage labels
for i, (senior, row) in enumerate(percentages.iterrows()):
    cumulative = 0
    for churn, value in row.items():
        ax.text(i, cumulative + value / 2, f"{value:.1f}%", ha='center', va='center', fontsize=10)
        cumulative += value

# Customization
plt.title("Churn by Senior Citizen (as %)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")
plt.xticks(ticks=[0, 1], labels=["No", "Yes"])
plt.legend(title="Churn")
plt.tight_layout()
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

# Columns to plot
columns_to_plot = [
    'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the figure and axes
num_plots = len(columns_to_plot)
fig, axes = plt.subplots(nrows=(num_plots + 2) // 3, ncols=3, figsize=(15, num_plots * 2.5))
axes = axes.flatten()

# Loop through columns and create a countplot for each
for i, column in enumerate(columns_to_plot):
    sns.countplot(data=df, x=column, ax=axes[i], palette="viridis",hue=df["Churn"])
    axes[i].set_title(f"Count of {column}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")

# Remove unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

# Customers with active phone services or multiple lines show lower churn compared to those without them.
# Fiber optic users tend to churn more compared to DSL or customers without internet service.
# Security features like online security, backups, and device protection are associated with reduced churn rates.
# Tech support and streaming services (TV and movies) users also show a noticeable churn trend when not subscribed.
